```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([80,95,90,95,100,105,110,115])
y=np.array([240,250,260,270,280,290,300,310])
```

```python
font1={'family':'arial','color':'darkred','size':20}
font2={'family':'serif','color':'blue','size':15}
```

```python
plt.plot(x,y)
plt.title("sports watch data", fontdict=font1 , loc='left')
plt.xlabel("average pulse", fontdict=font2, loc='left')
plt.ylabel("caloes burnage", fontdict=font2)
plt.show()
```
