```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([80,95,90,95,100,105,110,115])
y=np.array([240,250,260,270,280,290,300,310])
```

```python
plt.plot(x,y,marker='^',ls='dashed',c='r',mfc='y',mec='c')
plt.title("sports watch data")
plt.xlabel("average pulse")
plt.ylabel("caloes burnage")
plt.show()
```
